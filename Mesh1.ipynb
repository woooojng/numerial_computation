B.2 Differentiation Becomes Finite Differences

```python
The program forward_diff.py, shown below, computes the derivative
of f(x) using the approximation f'(x) = (f(x+h) -f(x))/h
```

```python
x = 1
h = 0.0001
sin = np.sin(x)

def diff(f, x, h):
    return (f(x+h) - f(x))/(h)
from math import *
import sys

approx_deriv = diff(sin, x, h)
exact = cos(x)
print ('The approximated value is:', approx_deriv)
print ('The correct value is: ', exact)
print ('The error is:', exact - approx_deriv)
```

```python
pi = math.pi
n = 1000
h = pi/(2*n)
x = linspace(0, pi/2, n+1)
s = np.sin(x)
z = zeros(len(s)) #len(list); number of component in the list
for i in range(len(z)-1):
    z[i] = (s[i+1] - s[i])/h
# Special formula for end point_
    z[-1] = (s[-1] - s[-2])/h
    plt.plot(x, z)
xfine = linspace(0, pi/2, 1001) # for more accurate plot
exact = np.cos(xfine)

plt.plot(xfine, exact)
plt.plot(xfine,z,'b',label='Approximate function')
plt.plot(xfine,exact,'r',label='Correct function')


#legend('Approximate function', 'Correct function')
plt.title('Approximate and discrete functions, n=%d' % n)
```
